```python
import numpy as np

X = np.array([
    [1, 1],
    [2, 1],
    [1, 2],
    [2, 2],
    [3, 4],
    [4, 3],
    [4, 5],
    [5, 4]
])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid()
plt.scatter(X[:, 0], X[:, 1])
```

```python
# 假设 初始分类点是(1, 2) (2, 1)
```

```python
centers = np.array([(1,1), (2, 1)])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid()
plt.scatter(X[:, 0], X[:, 1], color='grey')
plt.scatter(centers[0,0], centers[0,1], marker='s', color = 'green')
plt.scatter(centers[1,0], centers[1,1], marker='v', color = 'red')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid()
plt.scatter(X[:, 0], X[:, 1], color='grey')
plt.plot([1.5, 1.5], [0, 6])
plt.scatter(centers[0,0], centers[0,1], marker='s', color = 'green')
plt.scatter(centers[1,0], centers[1,1], marker='v', color = 'red')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid()
green = [0, 2]
red = [1, 3, 4, 5, 6, 7]

plt.scatter(X[green][:, 0], X[green][:, 1], color='green')
plt.scatter(X[red][:, 0], X[red][:, 1], color='red')

plt.plot([1.5, 1.5], [0, 6])

plt.scatter(centers[0,0], centers[0,1], marker='s', color = 'green')
plt.scatter(centers[1,0], centers[1,1], marker='v', color = 'red')
```
