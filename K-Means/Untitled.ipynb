```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
X = X[:, 2:]
y = iris.target
```

```python
def distance(M, N):
    return (np.sum((M - N) ** 2, axis = 1))** 0.5
```

```python
import random
            
def KMeans(X, k, n_iters):
    k = 3
    init_k_index = [random.randint(0, X.shape[0]) for i in np.arange(k)]
    center = np.array(X[init_k_index])
    nearest = np.array([(np.min(distance(center, item)) == distance(center, item)).tolist() for item in X])
    dots_index = [[] for i in np.arange(k)]
    dots = [[] for i in np.arange(k)]
    last_center = center
    while n_iters > 0:
        last_center = center
        for m in np.arange(k):
            for i, j in enumerate(nearest[:, m]):
                if j:
                    dots_index[m].append(i)
                    dots[m].append(X[i].tolist())
        
        
        for m in np.arange(k):
            center[m] = np.average(np.array(dots[m]), axis=0)
    
        if n_iters < 0:
            print('not enough n_iters')
        n_iters = n_iters - 1
    return center, dots

plt.figure(figsize=(18, 6))
for i in np.arange(3):
    plt.subplot(1, 3, i + 1)
    center = KMeans(X, 3, 40)
    plt.scatter(X[:, 0][y==0], X[:, 1][y==0])
    plt.scatter(X[:, 0][y==1], X[:, 1][y==1])
    plt.scatter(X[:, 0][y==2], X[:, 1][y==2])
    plt.scatter(center[0][0], center[0][1])
    plt.scatter(center[1][0], center[1][1])
    plt.scatter(center[2][0], center[2][1])
plt.show()
```
