```python
from sklearn import model_selection
```

```python
columns = train_data.columns.tolist()
len(columns)
```

```python
feat_cols = columns[:80]
```

```python
target = columns[80]
```

```python
string_list = []
number_list = []
for i in range(80):
    
    if (train_data[column[i]].dtype == 'object'):
        string_list.append(column[i])
    else:
        number_list.append(column[i])
print(string_list)
print(number_list)
```

```python
train_data[string_list]
```

```python
train_data[number_list]
```

```python
train_data[columns].dtype == 'object'
```

```python
from sklearn.preprocessing import LabelEncoder
for key in string_list:
    encoder= LabelEncoder().fit(train_data[key])
    train_data[key] = encoder.transform(train_data[key])
```

```python
train_data
```

```python
train_x, test_x, train_y, test_y = model_selection.train_test_split(train_data[feat_cols].values,
                                                                    train_data[target].values,
                                                                    random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_x, train_y)
model.score(train_x, train_y)
```

```python
y_pred = model.predict(test_x)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(test_y, y_pred)
```
