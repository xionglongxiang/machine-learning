```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 4, 0.1)
y = x ** 2 - 4 * x + 4
```

```python
plt.plot(x, y)
plt.show()
```

```python
def Y(x):
    return x ** 2 - 4 * x + 4
def d_Y(x):
    return 2 * x - 4
```

```python
theta = 0
```

```python
x_history = []
# eta 步长
def gradient_descent(x_init, eta, n_iters = 1e4, epsilon = 1e-8):
    _x = x_init
    i_iter = 0
    while i_iter < n_iters:
        last_x = _x
        x_history.append(_x)
        _x = _x - d_Y(_x) * eta
        if abs(Y(last_x) - Y(_x)) < epsilon:
            break
        i_iter = i_iter + 1
    return x
```

```python
x_history = []
gradient_descent(0, 1e-4)
plt.plot(x, y)
plt.plot(np.array(x_history), Y(np.array(x_history)), color="r")
plt.show()
```

```python
x_history = []
gradient_descent(0, 1.005)
plt.axis([-4, 8, 0, 20])
plt.plot(x, y)

plt.plot(np.array(x_history), Y(np.array(x_history)), color="r")
plt.show()
```
